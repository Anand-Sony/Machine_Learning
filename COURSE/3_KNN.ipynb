```python
# Define price ranges
bins = [0, df['price'].quantile(0.33), df['price'].quantile(0.66), float('inf')]
labels = ['Low', 'Medium', 'High']  # Labels for the price ranges

# Create a new column with price range categories
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels)

# Split data using the new target variable 'price_range'
x = df.iloc[:, 1:-1].values  # Exclude 'price' and 'price_range' from features
y = df['price_range'].values  # Use 'price_range' as target variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
import statsmodels.api as sn
x_cons = sn.add_constant(x)
```

```python
model = sn.OLS(y, x_cons).fit()
prediction = model.predict(x_cons)
```

```python
new_data = df.iloc[0:5, 1:]
new_data_cons = sn.add_constant(new_data)  # Add constant to new data
predictions = model.predict(new_data_cons)  # Predict on new data
```
