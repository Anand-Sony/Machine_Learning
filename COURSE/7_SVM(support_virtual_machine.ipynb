# **Classification Model :**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_classification.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Missing Value Imputation :

```python
df["Time_taken"].mean()
```

```python
df["Time_taken"].fillna(value = df["Time_taken"].mean(), inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Dummy variable Creation :

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=["3D_available", "Genre"], drop_first=True)
```

```python
df.head()
```

### X-Y Split :

```python
x = df.loc[:, df.columns!="Start_Tech_Oscar"]
```

```python
y = df["Start_Tech_Oscar"]
```

### Train-Test_Split :

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

### Standardizing Dataset :

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(x_train)
```

```python
x_train_std  = sc.transform(x_train)
```

```python
x_test_std = sc.transform(x_test)
```

```python
x_test_std
```

### Training Classification Tree :

```python
from sklearn import svm
```

```python
clf_svm_1 = svm.SVC(kernel = "linear", C=0.01)
clf_svm_1.fit(x_train_std, y_train)
```

### predicting Value :

```python
y_train_pred = clf_svm_1.predict(x_train_std)
y_test_pred = clf_svm_1.predict(x_test_std)
```

```python
y_test_pred
```

### Model Performance :

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
clf_svm_1.n_support_  # this will give me the no. of support vector
# first number will give no. of support vector for first class and second no. = no. of support vector for second class

# we have 2 class "0" and "1"
```

# **Grid Search :**

After changing c value for each model and take out result will be lengthy process

so we will use grid Search option of a scale

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C':(0.001, 0.005 , 0.01, 0.05, 0.1, 0.5 ,1 , 5 , 10, 50 , 100, 500, 1000 )}
```

```python
clf_svm_1 = svm.SVC(kernel="linear")
```

```python
svm_grid_lin = GridSearchCV(clf_svm_1, params, n_jobs=-1, cv = 10, verbose=1, scoring="accuracy")
```

n_jobs=-1 is used for maximize the performance of the computer

cv = cross-validation : fitting 10 fold for each 13 case 0.001,0.005, ..... ,500,1000

verbose is amount of message we are going to get while we are training the model



```python
svm_grid_lin.fit(x_train_std, y_train)
```

```python
svm_grid_lin.best_params_  # this will give me the best model out of 13
```

```python
linsvm_clf = svm_grid_lin.best_estimator_  # we will get parameter value according to our best model
```

```python
accuracy_score(y_test, linsvm_clf.predict(x_test_std))
```

### don't worry about it, if we have a large dataset gridSearch will definitely give better result.
