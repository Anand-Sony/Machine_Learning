```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/House_Price.csv", header=0)
```

```python
df.head()
```

```python
df["airport"] = df["airport"].replace({"YES":1, "NO":0})
```

```python
df = df.drop(['crime_rate', "dist1",	"dist2",	"dist3",	"dist4","waterbody","bus_ter"], axis=1)
```

```python
df.head()
```

```python
x = df[["price"]]
```

```python
y = df["airport"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_    #b0
```

```python
lr.intercept_    #b1
```

so this model do not use constant term means bo = 0 bydefault

so add constant term into our model

so first add constant to dependent variable

```python
import statsmodels.api as sn
```

```python
x_cons = sn.add_constant(x)
```

```python
x_cons.head()
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit = sm.Logit(y, x_cons).fit()
```

```python
logit.summary()
```
