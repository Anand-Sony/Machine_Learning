```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv('Movie_classification.csv', header=0)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

### Missing Value Imputation :

```python
df["Time_taken"].mean()
```

```python
df["Time_taken"].fillna(value= df["Time_taken"].mean() , inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Dummy Variable Creation :

```python
df.head()
```

```python
df = pd.get_dummies(df , columns=["3D_available", "Genre"], drop_first = True)
```

```python
df.head()
```

### X-Y Split :

```python
x = df.loc[:, df.columns!= "Start_Tech_Oscar"]
```

```python
type(x)
```

```python
x.head()
```

```python
x.shape
```

```python
y = df["Start_Tech_Oscar"]
type(y)
```

```python
y.head()
```

```python
y.shape
```

### Test Train Split :

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2, random_state=0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

### Training Classification Tree :

```python
from sklearn import tree
```

```python
clfTree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clfTree.fit(x_train, y_train)
```

### Predict Value Using trained model :

```python
y_train_pred = clfTree.predict(x_train)
```

```python
y_test_pred = clfTree.predict(x_test)
```

```python
y_test_pred
```

### Model Performance :

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

### Plotting Decision Tree :

```python
dot_data = tree.export_graphviz(clfTree, out_file=None, feature_names=x_train.columns, filled=True )
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

So on the parent item : 186 represent "0" and 218 are "1"

### Controlling Tree Growth :

```python
clfTree2 = tree.DecisionTreeClassifier(min_samples_leaf = 20, max_depth=4)
clfTree2.fit(x_train , y_train)
dot_data = tree.export_graphviz(clfTree2, out_file = None , feature_names = x_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

```python
accuracy_score(y_test, clfTree.predict(x_test))
```

```python
accuracy_score(y_test, clfTree2.predict(x_test))
```

## Bagging :

```python
from sklearn import tree
clf_tree  = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(estimator=clf_tree, n_estimators=1000, bootstrap= True, n_jobs=-1, random_state=42)
```

n_estimators =100 means we need 1000 trees to proceed in our model

and n_jobs=-1 means we need full processing power of our machine or computer

```python
bag_clf.fit(x_train, y_train)
```

```python
confusion_matrix(y_train, bag_clf.predict(x_train))
```

```python
confusion_matrix(y_test, bag_clf.predict(x_test))
```

```python
accuracy_score(y_test, bag_clf.predict(x_test))
```

# **Grid Search :**

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(n_estimators=250, random_state=42)
```

```python
param_grid = {"max_features": [4,5,6,7,8,9,10], "min_samples_split":[2,3,10]}
```

max_features is sqrt of number of the variables which comes out to be 4 and

there were 3 stopping conditions max_depth, min no of obs. at load and min no. of obs. at leaf node.

```python
grid_search = GridSearchCV(rf_clf, param_grid, n_jobs=-1, cv=5, scoring="accuracy")
```

cv means cross-validation

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

so we are getting best result at 'max_features': 6, 'min_samples_split': 3

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test, cvrf_clf.predict(x_test))
```

```python
confusion_matrix(y_test, cvrf_clf.predict(x_test))
```
