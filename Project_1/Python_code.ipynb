```python
x = dataset.iloc[: , :-1].values
y= dataset.iloc[: , -1].values
```

```python
from sklearn.model_selection import train_test_split   #Splitting the dataset into the training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = True)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(.90)
x_pca = pca.fit_transform(x_scaled)
x_pca
```

```python
pca.explained_variance_ratio_  # these are principle componenets
```

```python
from sklearn.model_selection import train_test_split   #Splitting the dataset into the training set and test set
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.3, random_state = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
new_data_point = [[0,148	,72	,35	,0	,33.6	,0.627,	50	]]
transformed_data_point = sc.transform(new_data_point)
predictions = classifier.predict(transformed_data_point)
print(predictions)
```

# Now ANN

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()    #initializes a new neural network
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))  # adds a fully connected layer with 10 neurons and ReLU(Rectified Linear Unit) helps introduce non-linearity into the model
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # adds a fully connected layer with 1 neuron and sigmoid activation function, typically used for binary classification tasks in neural networks.
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size = 32, epochs = 200)
```

```python
print(ann.predict(sc.transform([[0,148	,72	,35	,0	,33.6	,0.627,	50]])))
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```
