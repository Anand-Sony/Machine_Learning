```python
# import libraries :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# now import the salary_data file by pandas library
dataset = pd.read_csv("salary_data.csv")
```

```python
dataset
```

```python
dataset.shape  # check the shape of the dataframe(like no of rows and column)
```

```python
dataset.isnull().sum()  # this will tell you that if there is some null value present or not.(So 0 means No)
```

```python
x = dataset.iloc[:,:-1]   # In this we have selected the Independent Variable
y = dataset.iloc[:,-1]    # this is the target value(Dependent Variable)
```

```python
print(x)
print("\n")
print(y)
```

```python
dataset.corr()  # this is used to identify the relation between the variable in the dataset
```

```python
# now Splitting the dataset into the Training set and Test set
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=2) #0.2 represent 20% data used for testing and 80% for training.
print(x_train)
```

```python
# Training the Simple Linear Regression model on the Training set
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Now Predicting the Test set results
```

```python
print("Dataset of y value =\n",y_test)
y_pred = reg.predict(x_test)
print("Predicted value = ",y_pred)
```

```python
coeff = reg.coef_
intercept = reg.intercept_
print("Coefficient = ",coeff)
print("Intercept = ",intercept)
```

```python
from sklearn.metrics import r2_score , accuracy_score
r_square = r2_score(y_test , y_pred)
print("Coeffecient of Determination = ",r_square)
```

```python
# now we will visualize the data for training set result
```

```python
plt.scatter(x_train , y_train , color ="green")
plt.plot(x_train , reg.predict(x_train) , color ="blue")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Salary Vs Experience(Training Set)")
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'green')
plt.plot(x_train, reg.predict(x_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
