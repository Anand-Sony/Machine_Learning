### **PCA reduces the number of variables or features in a data set while still preserving the most important information like major trends or patterns.**

### First we will import the libraries : and then the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
```

```python
dataset
```

Now check the null values are present or not in the dataset :

```python
dataset.isnull().sum()
```

Now extracts the values of a dataset into a variable x in Python using the pandas library.

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler #scaling the data using StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled  #  each feature has been centered around 0 and scaled to have unit variance.
```

### StandardScaler standardizes features by removing the mean and scaling to unit variance.

# Now **Principal Component Analysis (PCA)**

```python
from sklearn.decomposition import PCA
pca = PCA(0.9)    #we can also use PCA(n_components = 2) : if we know that 2 component are contrubuting major portion of the dataset.
x_pca = pca.fit_transform(x_scaled)
x_pca
```

```python
pca.explained_variance_ratio_  # these are principle componenets
```

### Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_scaled,y ,test_size=0.35 , random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train , y_train)
# fits the k-Nearest Neighbors classifier to the training data x_train and the corresponding target labels y_train.
```

```python
y_pred = knn.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

**Predicting the New Data**

```python
new_data_point = [[0.2, 1, 6, 0.8]]
transformed_data_point = sc.transform(new_data_point)
pred = knn.predict(transformed_data_point)
print(pred)
```

# Confusion Matrix

**confusion matrix summarizes the performance of a classification model by showing the number of correct and incorrect predictions for each class.**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
